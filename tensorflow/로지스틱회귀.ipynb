### 로지스틱 회귀

로지스틱 회귀는 시그모이드 함수를 활용하여 2진분류 수행<br>
<br>
$ y = {1 \over 1+e^{-(ax+b)}} $ <br>
<br>
실제 값이 1 or 0 인가에 따라 사용되는 오차함수가 다른데<br>
실제 값이 1일 경우: $-log(h)$,<br>
실제 값이 0일 경우: $-log(1-h)$ 를 사용하여 오차를 구한다.<br>
<br>
오차 함수: $ f(h) = -{1 \over n}\sum y log(h) + (1-y) log(1-h) $<br>
<br>
위의 공식을 기울기 'a' 와 절편 'b'로 각 편미분한 식은 다음과 같다.

$$ { \partial f(a,b) \over \partial{a} } = (sigmoid(ax+b)-y)x$$
$$ { \partial f(a,b) \over \partial{b} } = sigmoid(ax+b)-y$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# -log(h), -log(1-h) 로그함수
x = np.linspace(0.0+1e-5, 1.0-1e-5, num = 100)
#print(x)
y = -np.log(x)
y1 = -np.log(1-x)
plt.plot(x,y, label="-log(h)")
plt.plot(x,y1, label="-log(1-h)")
plt.legend()
plt.show()
```

```python
x = np.array([2,4,6,8,10,12,14])
y = np.array([0, 0, 0, 1, 1, 1, 1])

plt.scatter(x, y)
plt.xlim(0, 15)
plt.ylim(-.1, 1.1)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
a = 0
b = 0
lr = 0.1
epoches = 2001

for ep in range(epoches):
    y_pred = sigmoid(a*x+b)
    err = -np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))
    print(f"err: {err:.4f}")
    a_diff = ((y_pred-y)*x).mean()
    b_diff = (y_pred-y).mean()

    a -= a_diff*lr
    b -= b_diff*lr

    #print(f"epoch:{ep}, a_diff:{a_diff:.3f}, b_diff:{b_diff:.3f}")
    print(f"epoch:{ep}, a:{a:.3f}, b:{b:.3f}")
```

```python
y_pred = sigmoid(a*x+b)
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlim(0, 15)
plt.ylim(-.1, 1.1)
plt.show()
```
