```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df_log = pd.read_csv("./data/deep_dqn.csv")
df_log.drop(columns=df_log.columns[0], axis=1, inplace=True)
df_ohlc = pd.read_csv("./data/KRW_BTC_traing.csv")
df_ohlc.head()
```

```python
df_buy = df_log[df_log.action == 0]
df_sell = df_log[df_log.action == 1]
```

```python
# 종가 그래프 그리기
fig = px.line(df_log, x='date', y='price')
fig.show()
```

```python
# 종가 그래프에 매수, 매도 마커 표시하기
#fig = go.Figure()
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(x=df_log.date, y=df_log.price, line=dict(width=1.1), mode='lines', name="price") , secondary_y=False)
fig.add_trace(go.Scatter(x=df_buy.date, y=df_buy.price+50000, mode='markers', marker_symbol=6, name='buy'), secondary_y=False)
fig.add_trace(go.Scatter(x=df_sell.date, y=df_sell.price-50000, mode='markers', marker_symbol=5, name='sell'), secondary_y=False)
fig.add_trace(go.Scatter(x=df_ohlc.date, y=df_ohlc.ma_7, mode='lines', line=dict(width=1), name='mva7'), secondary_y=False)
fig.add_trace(go.Scatter(x=df_ohlc.date, y=df_ohlc.ma_15, mode='lines', line=dict(width=1), name='mva15'), secondary_y=False)
fig.add_trace(go.Scatter(x=df_ohlc.date, y=df_ohlc.ma_30, mode='lines', line=dict(width=1), name='mva30'), secondary_y=False)
fig.add_trace(go.Scatter(x=df_log.date, y=df_log.profitloss, mode='lines', line=dict(width=1), name='profitloss'), secondary_y=True)
fig.show()
```
