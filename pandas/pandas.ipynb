```python
# 리스트로 구성된 데이터 클래스를 데이터 프레임으로 만들기

from dataclasses import dataclass

@dataclass
class Agent_info:    
    data_idx: int = -1
    balance: float=0.

@dataclass
class Trading_info:
    data_idx: int = -1
    action: int = -1
     

list_agent = []
list_agent.append(Agent_info(0, 10))
list_agent.append(Agent_info(1, 11))
list_agent.append(Agent_info(2, 12))

list_trading = []
list_trading.append(Trading_info(0, 0))
list_trading.append(Trading_info(1, 1))
list_trading.append(Trading_info(2, 0))

df = pd.DataFrame(list_agent)
df1 = pd.DataFrame(list_trading)
```
